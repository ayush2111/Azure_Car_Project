# DATA READING


```python
df = spark.read.format('parquet')\
        .option("header", "true")\
        .load('abfss://silver@carstorageaccount.dfs.core.windows.net/rawdata')
```

```python
display(df)
```

# DATA TRANSFORMATION

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = df.withColumn('model_category',split(col('Model_ID'),'-')[0])
```

```python
df.withColumn('RevenuPerUnit',col('Revenue')/col('Units_Sold'))
```

# AD-HOC

```python
df.groupBy('Year','BranchName').agg(sum('Units_Sold').alias('Total_Units')).sort('Year','Total_Units',ascending = [1,0])
```

# DATA WRITING

```python
df.write.format('parquet').mode('overwrite').save('abfss://silver@carstorageaccount.dfs.core.windows.net/carsales').save()
```
